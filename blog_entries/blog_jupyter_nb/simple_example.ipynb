```python
import geopandas as gpd

df = gpd.read_file('../../data/mexico_cities_simplemaps.csv',  encoding='utf8')
for col in ['lat', 'lng']:
    df[col] = df[col].astype('float')
```

```python
df = df.set_geometry(gpd.points_from_xy(x=df['lng'], y=df['lat'], crs=4326))
```
